```python
import pandas as pd

df = pd.read_pickle('OzoneISA_topic_clustering_export.pkl')
df
```

```python
# dataframe of topic numbers and associated top words
# these are the top 5 ranked important words for each topic
top_words = df[['category', 'top_words']].drop_duplicates().set_index('category')
top_words
```

```python
# set plotting variables
x = df['dim_0']
y = df['dim_1']
color = df['category']
labels = df['top_words']
included = df['included']
```

```python
# TSNE representation of topic clustering, colored by topic

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,12))

#cmap = sns.color_palette('hls', 20)

ax = sns.scatterplot(x=x, y=y,
                     hue=color,
                     palette='jet',
                     alpha = 0.4,
                    legend='full'
                    ).set(xlabel="", ylabel = "")
```

```python
# TSNE representation of topic clustering, colored by inclusion/exclusion in final draft ISA

plt.figure(figsize=(16,12))

ax = sns.scatterplot(x=x, y=y,
                     hue=included,
                     hue_order=[False, True],
                     palette=['lightgray','black'],
                    legend='full',
                     alpha = 0.5
                    ).set(xlabel="", ylabel = "")
```
